```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import re
```

```python
df = pd.read_excel('BigV All Time Stats (SportsTG) for Milestones Project v2 (Coyne mins, 48 factor).xlsx')
```

```python
# this a pattern to detects hyphens specifically
hyphen_pattern = re.compile(r"-")

# finding names with hyphens
first_name_hyphen = df[df['First Name'].apply(lambda x: bool(hyphen_pattern.search(str(x))))]
family_name_hyphen = df[df['Family Name'].apply(lambda x: bool(hyphen_pattern.search(str(x))))]


names_with_hyphens = pd.concat([first_name_hyphen, family_name_hyphen]).drop_duplicates()
```

```python

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(names_with_hyphens)
```

```python
# Checking for names that have "Non Member" first or last
non_member_first = df[df['First Name'].str.contains("Non Member", case=False, na=False)]
non_member_last = df[df['Family Name'].str.contains("Non Member", case=False, na=False)]


non_member_names = pd.concat([non_member_first, non_member_last]).drop_duplicates()

non_member_names
```

```python
# Creating a cleaned version of the dataset


# Removing the " - Non Member" part from the 'First Name' field
df['First Name'] = df['First Name'].str.replace(r"\s*-\s*Non Member", "", case=False, regex=True)

# Re checking entries old cells with "Non Member" to check theyve been cleaned
cleaned_non_member_first = df[df['First Name'].str.contains("JAYCE", case=False, na=False)]
cleaned_non_member_other = df[df['First Name'].str.contains("CLAIRE", case=False, na=False)]
cleaned_non_member_names = pd.concat([cleaned_non_member_first, cleaned_non_member_other]).drop_duplicates()
```

```python
# Checking for missing or empty values in first/last name
missing_names = df[
    df['First Name'].isna() | df['Family Name'].isna() |
    (df['First Name'].str.strip() == "") | (df['Family Name'].str.strip() == "")
]
```

```python
# Remove those rows as we dont know the player
df = df[
    df['First Name'].notna() & df['Family Name'].notna() &
    (df['First Name'].str.strip() != "") & (df['Family Name'].str.strip() != "")
]
```

```python
# making sure all first and last names start with capital letter and rest are lowercase
df["First Name"] = df["First Name"].str.title()
df["Family Name"] = df["Family Name"].str.title()

# show a couple outputs to check
df[["First Name", "Family Name"]].head()
```

```python
# Remove rows where '*' shows up in first or last name
df = df[~df['First Name'].str.contains(r"\*", na=False) &
        ~df['Family Name'].str.contains(r"\*", na=False)]
```

```python
# join first and last name together into a full name
df['Name'] = df['First Name'].str.strip() + " " + df['Family Name'].str.strip()


columns = ['Name'] + [col for col in df.columns if col not in ['First Name', 'Family Name', 'Name']]
df = df[columns]
```

```python
# add back gender column as it accidently got removed 
gender_info = df.groupby("Name")["Gender"].first().reset_index()

# Sum all stat columns per player
stat_columns = df.columns[df.columns.get_loc("GP"):]
dftotalscareer = df.groupby("Name")[stat_columns].sum().reset_index()
```

```python
# Create a copy of the original cleaned data
dfavgsbs = df.copy()

# List of columns to average starting from MIN to FTA
columns_to_avg = dfavgsbs.columns[dfavgsbs.columns.get_loc("MIN"):]

# Dividing each of these columns by GP to get per game averages
for col in columns_to_avg:
    dfavgsbs[col] = dfavgsbs[col] / dfavgsbs["GP"]
    # 1 decimal place
dfavgsbs[columns_to_avg] = dfavgsbs[columns_to_avg].round(1)
```

```python
dfavgsbs
```

```python


# Copy the career totals dataset
dfavgcareer = dftotalscareer.copy()

# Get the list of columns to average except name and gp
cols_to_avg = dfavgcareer.columns.difference(['Name', 'GP'])

# Divide each stat column by total games played to get per-game career averages
for col in cols_to_avg:
    dfavgcareer[col] = dfavgcareer[col] / dfavgcareer['GP']

# Round to 1 decimal places
dfavgcareer[cols_to_avg] = dfavgcareer[cols_to_avg].round(1)
```

```python
# Start from the original cleaned dataset
dfscaledavgsbs = df.copy()


columns_to_rescale = dfscaledavgsbs.columns[dfscaledavgsbs.columns.get_loc("MIN"):]

#  (Stat x Factor) / GP
for col in columns_to_rescale:
    dfscaledavgsbs[col] = (dfscaledavgsbs[col] * dfscaledavgsbs["Factor"]) / dfscaledavgsbs["GP"]


dfscaledavgsbs[columns_to_rescale] = dfscaledavgsbs[columns_to_rescale].round(1)
```

```python
dfscaledavgsbs
```

```python
# Start from the original career totals dataset
dfscaledavgcareer = dftotalscareer.copy()

# scaling major stat columns only
columns_to_rescale = dfscaledavgcareer.columns[dfscaledavgcareer.columns.get_loc("MIN"):]

# Merge the average factor per player from original df 
avg_factors = df.groupby("Name")["Factor"].mean().reset_index()
dfscaledavgcareer = dfscaledavgcareer.merge(avg_factors, on="Name", how="left")

#  (Career Total Stat * Average Factor) / Total Games Played
for col in columns_to_rescale:
    dfscaledavgcareer[col] = (dfscaledavgcareer[col] * dfscaledavgcareer["Factor"]) / dfscaledavgcareer["GP"]

# Drop the Factor column now thatwe used it
dfscaledavgcareer.drop(columns=["Factor"], inplace=True)


dfscaledavgcareer[columns_to_rescale] = dfscaledavgcareer[columns_to_rescale].round(1)
```

```python
# Start from the per-game averages dataset 
dfscaled30avgsbs = dfavgsbs.copy()


columns_to_rescale_30 = dfscaled30avgsbs.columns[dfscaled30avgsbs.columns.get_loc("PTS"):]

# (stat / MIN) * 30
for col in columns_to_rescale_30:
    dfscaled30avgsbs[col] = (dfscaled30avgsbs[col] / dfscaled30avgsbs["MIN"]) * 30


dfscaled30avgsbs[columns_to_rescale_30] = dfscaled30avgsbs[columns_to_rescale_30].round(1)
```

```python
# Start from the career averages dataset 
dfscaled30avgcareer = dfavgcareer.copy()


columns_to_rescale_30_career = dfscaled30avgcareer.columns[dfscaled30avgcareer.columns.get_loc("PTS"):]

#  (Stat / MIN) * 30
for col in columns_to_rescale_30_career:
    dfscaled30avgcareer[col] = (dfscaled30avgcareer[col] / dfscaled30avgcareer["MIN"]) * 30


dfscaled30avgcareer[columns_to_rescale_30_career] = dfscaled30avgcareer[columns_to_rescale_30_career].round(1)
```

```python
#now we add the shooting percentages for each dataset as requested later on by industry

def add_shooting_percentages_with_fg(df):
    df = df.copy()
    # Calculate raw percentages
    df["2P%"] = (df["2PM"] / df["2PA"] * 100).replace([np.inf, -np.inf], np.nan)
    df["3P%"] = (df["3PM"] / df["3PA"] * 100).replace([np.inf, -np.inf], np.nan)
    df["FT%"] = (df["FTM"] / df["FTA"] * 100).replace([np.inf, -np.inf], np.nan)
    df["FG%"] = (df["FGM"] / df["FGA"] * 100).replace([np.inf, -np.inf], np.nan)

    # force set to 0 if 0 were made
    df.loc[(df["2PM"] == 0) & (df["2PA"] == 0), "2P%"] = 0.0
    df.loc[(df["3PM"] == 0) & (df["3PA"] == 0), "3P%"] = 0.0
    df.loc[(df["FTM"] == 0) & (df["FTA"] == 0), "FT%"] = 0.0
    df.loc[(df["FGM"] == 0) & (df["FGA"] == 0), "FG%"] = 0.0

    # Round to one decimal place
    df["2P%"] = df["2P%"].round(1)
    df["3P%"] = df["3P%"].round(1)
    df["FT%"] = df["FT%"].round(1)
    df["FG%"] = df["FG%"].round(1)

    # Reordered the columns
    cols = df.columns.tolist()
    for made, att, pct in [("FGM", "FGA", "FG%"), ("2PM", "2PA", "2P%"), ("3PM", "3PA", "3P%"), ("FTM", "FTA", "FT%")]:
        if pct in cols:
            cols.remove(pct)
            att_index = cols.index(att)
            cols.insert(att_index + 1, pct)

    return df[cols]



# Apply updated functions to all the datasets
df = add_shooting_percentages_with_fg(df)
dftotalscareer = add_shooting_percentages_with_fg(dftotalscareer)
dfavgsbs = add_shooting_percentages_with_fg(dfavgsbs)
dfavgcareer = add_shooting_percentages_with_fg(dfavgcareer)
dfscaledavgsbs = add_shooting_percentages_with_fg(dfscaledavgsbs)
dfscaledavgcareer = add_shooting_percentages_with_fg(dfscaledavgcareer)
dfscaled30avgsbs = add_shooting_percentages_with_fg(dfscaled30avgsbs)
dfscaled30avgcareer = add_shooting_percentages_with_fg(dfscaled30avgcareer)
```
